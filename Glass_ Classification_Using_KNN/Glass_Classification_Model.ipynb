9. Ba: Barium
10. Fe: Iron

```python
#dividing the dataset to X and y 
X=df.drop(columns=('Type'))
y=df['Type']
```

```python
#Incializing Standard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_X=scaler.fit_transform(X)
```

```python
scaled_X
```

```python
#converting back to a dataframe
scaled_X=pd.DataFrame(scaled_X,columns=X.columns)
```

```python
scaled_X
```

```python
#spliting the dataset to train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# finding the best k to work with in the KNN the K with the highest accuracy score 
accuracies=[]
k_values=range(1,30)
for k in k_values:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    y_pred=knn.predict(X_test)
    acc=accuracy_score(y_test,y_pred)
    accuracies.append(acc)
```

```python
plt.figure(figsize=(12,8))
plt.plot(k_values,accuracies,marker='o')
plt.title('KNN Accuracy for Different K Values')
plt.xlabel('k (Number of Neighbors)')
plt.ylabel('K_Accuracy')
plt.xticks(k_values)
plt.show()
```

```python
#Training the model with k_value with the highest accuracy k=1
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
```

```python
print('Accuracy:', accuracy)
```

```python
# Classification Report
report=classification_report(y_test,y_pred)
print('Classification Report:\n', report)
```

```python
## Confusion Matrix
cm=confusion_matrix(y_test,y_pred)
print('Confusion Matrix :\n', cm)
```
